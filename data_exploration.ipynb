```python
import pandas as pd
import numpy as np
import os
from sklearn.impute import KNNImputer
from typing import List, Tuple
```

```python
input_directory = r"C:\Users\JacopoBinati\OneDrive - Venionaire Capital\Desktop\python codes\multiples\to be cleaned\output"

def merge_csv_files(input_dir):
    dataframes = []
    
    for file_name in os.listdir(input_dir):
        if file_name.endswith('.csv'):
            file_path = os.path.join(input_dir, file_name)
            print(f"Reading file: {file_name}")
            try:
                df = pd.read_csv(file_path)
                dataframes.append(df)
            except Exception as e:
                print(f"Error reading file {file_name}: {e}")
    
    if dataframes:
        data = pd.concat(dataframes, ignore_index=True)
        return data
    else:
        print("No valid CSV files found for merging.")
        return pd.DataFrame()

data = merge_csv_files(input_directory)
```
